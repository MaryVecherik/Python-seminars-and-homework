```python
# Дана функция f(x) = 5x^2 + 10x - 30

# Определить корни (найти такие х, при которых f(x) = 0) - корни уравнения f(x) = 0

# Найти интервалы, на которых функция возрастает (решить неравенства где производная больше или меньше нуля)

# Найти интервалы, на которых функция убывает

# Построить график

# Вычислить вершину (найти производню и приравнять к нулю)

# Определить промежутки, на котором f > 0 (Найти такие x, при которых f(х) > 0) неравенствo f(x) > 0

# Определить промежутки, на котором f < 0 ( Найти такие x, при которых f(х) < 0) неравенствo f(x) < 0

# Какие методы потребуются: (sympy)
# 1) Методы решающие уравнения
# 2) Методы решающие неравенства
# 3) Методы, которые находят производную
# 4) Метод, для простроения графика
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f = 5*x**2 + 10*x - 30
```

```python
print('Корни уравнения')
solve(f, x)
```

```python
#вариант 1
# print('Промежутки, на котором f > 0')
# print(solve(f>0, x))
# print('Промежутки, на котором f < 0')
# print(solve(f<0, x))

#вариант 2
print('Промежуток, на котором f > 0')
pprint(solveset(f>0, x, domain=S.Reals), use_unicode=True)
print('Промежуток, на котором f < 0')
pprint(solveset(f<0, x, domain=S.Reals), use_unicode=True)
```

```python
#вариант 1
print('График функции')
plot(f)
f
```

```python
#вариант 2
import matplotlib.pyplot as plt

print('График функции')
px = [i for i in range(-10, 10)]
fx = lambda x: 5*x**2 + 10*x - 30
py = list(map(fx, range(-10, 10)))
plt.plot(px, py, 'r')
plt.show()
```

```python
print('Производная функции')
#вариант 1
# p = f.diff(x)
# print(p)

#вариант 2
diff_f = diff(f, x)
print(diff_f)
```

```python
print('Вершина')
solve(diff_f, x)
```

```python
#вариант 1
# print('Интервалы, на которых функция возрастает')
# print(solve(p>0, x))
# print('Интервалы, на которых функция убывает')
# print(solve(p<0, x))

#вариант 2
print('Интервалы, на которых функция возрастает')
pprint(solveset(diff_f>0, x, domain=S.Reals), use_unicode=True)
print('Интервалы, на которых функция убывает')
pprint(solveset(diff_f<0, x, domain=S.Reals), use_unicode=True)
```
