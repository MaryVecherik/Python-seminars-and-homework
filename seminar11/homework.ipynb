```python
# Дана функция f(x) = sin(x)^2 - cos(x)^2

# Определить корни
# Найти интервалы, на которых функция возрастает
# Найти интервалы, на которых функция убывает
# Построить график
# Вычислить вершину
# Определить промежутки, на котором f > 0
# Определить промежутки, на котором f < 0
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
```

```python
print('Корни уравнения')
solve(f, x)
```

```python
ans = solve(f, x)
for i in ans:
    print(i.evalf())
```

```python
#вариант 1
# print('Промежутки, на котором f > 0')
# print(solve(f>0, x))
# print('Промежутки, на котором f < 0')
# print(solve(f<0, x))

#вариант 2
print('Промежуток, на котором f > 0')
pprint(solveset(f>0, x, domain=S.Reals), use_unicode=True)
print('Промежуток, на котором f < 0')
pprint(solveset(f<0, x, domain=S.Reals), use_unicode=True)
```

```python
#вариант 1
print('График функции')
plot(f)
f
```

```python
#вариант 2
import matplotlib.pyplot as plt

print('График функции')
px = [i for i in range(-10, 10)]
fx = lambda x: sin(x)**2 - cos(x)**2
py = list(map(fx, range(-10, 10)))
plt.plot(px, py, 'b')
plt.show()
```

```python
print('Производная функции')
p = f.diff(x)
print(p)
```

```python
print('Вершина')
solve(p, x)
```

```python
#вариант 1
print('Интервалы, на которых функция возрастает')
print(solve(p>0, x))
print('Интервалы, на которых функция убывает')
print(solve(p<0, x))

#вариант 2
print('Интервалы, на которых функция возрастает')
pprint(solveset(p>0, x, domain=S.Reals), use_unicode=True)
print('Интервалы, на которых функция убывает')
pprint(solveset(p<0, x, domain=S.Reals), use_unicode=True)
```
